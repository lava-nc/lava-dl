```python
from piecewise_linear_silu import PiecewiseLinearSiLU
from torch import nn
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.arange(-10, 10, 0.1).float()
```

```python
silu = nn.SiLU()
pl_silu = PiecewiseLinearSiLU()
```

```python
y_silu = silu(x)
y_pl_silu = pl_silu(x)
```

```python
plt.plot(x, y_silu.numpy())
plt.plot(x, y_pl_silu.numpy())
plt.plot(x, torch.abs(y_silu - y_pl_silu).numpy())
```
